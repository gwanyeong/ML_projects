```python
import os
from pymatgen.io.vasp import Vasprun
from ase.io import read, write
from ase.io import Trajectory
from IPython.display import Image
```

```python
def createFolder(directory):
    try:
        if not os.path.exists(directory):
            os.makedirs(directory)
    except OSError:
        print('Error: Creating directory. ' + directory)
```

```python
v = Vasprun('vasprun.xml')

images = []
createFolder('images')
for i in range(len(v.ionic_steps)):
    struc = v.ionic_steps[i]['structure']
    struc.to(fmt = 'cif', filename = 'images/%03d.cif' % i)
    image = read('images/%03d.cif' % i)
    images.append(image)
    
write(filename = 'images.traj', images = images)
traj = Trajectory('images.traj')
write('images_1.gif', traj, rotation = ("-90x, 30y, 0z")) # 
write('images_2.gif', traj) # 
print('')
```
